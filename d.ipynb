```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('d.csv')
```

```python
df = df.applymap(lambda x: round(x,4))
```

```python
fig,ax = plt.subplots(figsize=(20,12))
_ = ax.plot(df.d,df.catch,label='% members tabled')
_ = ax.plot(df.d,df.share,label='% guests shared across tables')
_ = ax.plot(df.d,df.collapse,label='% guest collapse')
_ = plt.axhline(0.8,c='grey')
_ = plt.axvline(0.0657,c='grey')
_ = ax.legend()
```
